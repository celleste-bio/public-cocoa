# **Import Libraries**

```python
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
```

# **Function to replace "-" into null + calculate the percentage of missing values in each coloumn**

```python
def replace_char_calculate_prc_missing_values(df,numeric_columns):
  numeric_columns = numeric_columns
  # List of all columns
  all_columns = df.columns
  # Identify non-numeric columns
  categorical_columns = [col for col in all_columns if col not in numeric_columns]
  # Replace "-" with NaN in both numerical and categorical columns
  df[numeric_columns] = df[numeric_columns].replace('-', np.nan)
  df[categorical_columns] = df[categorical_columns].replace('-', np.nan)
  # Convert numerical columns to float
  df[numeric_columns] = df[numeric_columns].astype(float)
  #-----------------------------------------------------
  missing_percentage = df.isnull().mean() * 100
  # Create a DataFrame to display the results
  missing_info = pd.DataFrame({
      'Column': missing_percentage.index,
      'Missing Percentage': missing_percentage.values })

  return missing_info.sort_values(by=['Missing Percentage'])
```

# **Function for bean.csv**

```python
def bean_missing_values(df,numeric_columns):
  numeric_columns = numeric_columns
  # List of all columns
  all_columns = df.columns
  # Identify non-numeric columns
  categorical_columns = [col for col in all_columns if col not in numeric_columns]
  # Replace "-" with NaN in both numerical and categorical columns
  #df[numeric_columns] = df[numeric_columns].replace('-', np.nan)
  #df[categorical_columns] = df[categorical_columns].replace('-', np.nan)
  # Convert numerical columns to float
  df[numeric_columns] = df[numeric_columns].astype(float)
  #-----------------------------------------------------
  missing_percentage = df.isnull().mean() * 100
  # Create a DataFrame to display the results
  missing_info = pd.DataFrame({
      'Column': missing_percentage.index,
      'Missing Percentage': missing_percentage.values })

  return missing_info.sort_values(by=['Missing Percentage'])
```

# **yield**

```python
df = pd.read_csv('/content/yield_v1.csv')
```

```python
print(replace_char_calculate_prc_missing_values(df,['Pod index', 'Yield ha', 'Yield tree', 'Seed index', 'Wet weight beans per tree', 'Yield efficiency']))
```

```python
missing_info = replace_char_calculate_prc_missing_values(df,['Pod index', 'Yield ha', 'Yield tree', 'Seed index', 'Wet weight beans per tree', 'Yield efficiency'])
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.bar(missing_info['Column'], missing_info['Missing Percentage'], color='skyblue')
plt.xlabel('Columns')
plt.ylabel('Missing Percentage')
plt.title('yield')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
